```python
import pandas as pd
import os

# Lire tous les fichiers Parquet du dossier
folder = "data/weather_lake"
files = [os.path.join(folder, f) for f in os.listdir(folder) if f.endswith(".parquet")]

df = pd.concat([pd.read_parquet(f) for f in files], ignore_index=True)

# Vérifier les colonnes
print("Colonnes : ", df.columns)

# Vérifier types et valeurs manquantes
print(df.info())

# Aperçu des données
print(df.head(10))

# Vérifier alertes de température
print(df[['city', 'temperature', 'alert_temp']].sample(10))
```
