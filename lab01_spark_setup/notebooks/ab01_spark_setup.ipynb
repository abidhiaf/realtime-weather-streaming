```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("LAB01_Spark_Setup")
    .master("local[*]")
    .getOrCreate()
)

spark
```

```python
data = [
    ("Alice", 24),
    ("Bob", 29),
    ("Céline", 31),
    ("David", 22),
]

df = spark.createDataFrame(data, schema=["nom", "age"])
df.show()
```

```python
from pyspark.sql.functions import col

df_filtre = df.filter(col("age") > 25)
df_filtre.show()
```

```python
from pyspark.sql.functions import avg

df_agg = df.agg(avg(col("age")).alias("age_moyen"))
df_agg.show()
```
